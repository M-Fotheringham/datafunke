## Probabilistic Zones of Regression

```python
import matplotlib.pyplot as plt
from itertools import product
from astropathdb import AstroDB
from datafunks import tdistogram, hist_plotter, get_cell_counts, get_area
```

### Axis Panel

```python
db = AstroDB(database="wsi02")
```

```python
# simple example of 50um density histogram
tdist = tdistogram(sampleid=102, database=db, t_hist_step=50, prob=False, tdist_filter=0, all_reg=True, excl_ln=True, samplewise=True, smoothed=False)
```

```python
tdist
```

```python
# Filter total density per lineage
total = tdist[tdist["exprphenotype"] == "Total"].reset_index(drop=True)
```

```python
# tdist_smoothed and y_smoothed for smoothed;
# tdist_bin prob for unsmoothed probability
# tdist_bin prop for unsmoothed lineage proportion
# I don't reformat the y labels inside the function for proportion, but works # for density and probability

hist_plotter(total, f"tdist_bin", f"density_mm", sample="test samples", pheno=["CD8", "FoxP3", "CD163", "FoxP3CD8"], save=False, prob=False)
```

```python
# For looping over different inputs
param = {
    "samples": [102, 114, 126, 194, 543, 546],
    
    "area": [("goodtissue", None, False), ("regression", 0, True)],
    
    "smooth": [(False, "unsmoothed", "bin", {"prob": "prob", "den": 
    "density_mm", "prop": "prop"}), (True, "smoothed", "smoothed", {"prob": "smoothed_y", 
    "den": "smoothed_y", "prop": "smoothed_y"})],
    
    "p": [(False, "den"), (True, "prob")],
    
    "markers": [(["CD8", "CD163"], "CD8, CD163"), (["CD8", "CD163", "FoxP3", "FoxP3CD8"], "Axis Immune Cell")]
    }
```

```python
# Takes like 4 min
for area, smooth, p, markers in product(
    param["area"], param["smooth"], param["p"], param["markers"]
):
    print(area[0], p[1], smooth[1], markers[1])
    
    bins = tdistogram(sampleid=param["samples"], database=db, t_hist_step=50, prob=p[0], tdist_filter=area[1], all_reg=area[2], excl_ln=True, samplewise=False, smoothed=smooth[0])

    total = bins[bins["exprphenotype"] == "Total"].reset_index(drop=True)

    hist_plotter(total, f"tdist_{smooth[2]}", f"{smooth[3][p[1]]}", sample="test samples", pheno=markers[0], save=False, prob=p[0])

    plt.savefig(f"histograms/{area[0]}/{smooth[1]}/testsamples_{smooth[1]}_{p[1]}_{markers[1]}_50um_hist.png", dpi=900, transparent=True, bbox_inches="tight")

    plt.cla()
    plt.clf()
```

```python
# simple example for testing that props look right
bins = tdistogram(sampleid=887, database=db, phenos=["CD8"], t_hist_step=50, prob=False, prop=True, excl_ln=True, samplewise=True, smoothed=True, altdb="wsi14")
```

```python
total = bins[bins["exprphenotype"].isin([0, 4, 8, 16, 32, 64, 80, 40, "Total"])].reset_index(drop=True)
total["exprphenotype"] = total["exprphenotype"].replace(exprdict)
```
